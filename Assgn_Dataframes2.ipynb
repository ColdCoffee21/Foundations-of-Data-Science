```
#Q13,14
empRates=pd.read_csv('https://ec.europa.eu/eurostat/estat-navtree-portlet-prod/BulkDownloadListing?file=data/tps00053.tsv.gz',compression='gzip',sep=',|\t')
empRates
```

```
#Q15
empRates['2008 '] = empRates['2008 '].str.extract('(\d+)').astype(float)
empRates['2009 '] = empRates['2009 '].str.extract('(\d+)').astype(float)
empRates['2010 '] = empRates['2010 '].str.extract('(\d+)').astype(float)
empRates['2011 '] = empRates['2011 '].str.extract('(\d+)').astype(float)
empRates['2013 '] = empRates['2013 '].str.extract('(\d+)').astype(float)
empRates['2014 '] = empRates['2014 '].str.extract('(\d+)').astype(float)
empRates['2015 '] = empRates['2015 '].str.extract('(\d+)').astype(float)
empRates['2016 '] = empRates['2016 '].str.extract('(\d+)').astype(float)
empRates['2017 '] = empRates['2017 '].str.extract('(\d+)').astype(float)
empRates['2018 '] = empRates['2018 '].str.extract('(\d+)').astype(float)
```

```
empRates.dtypes
```
