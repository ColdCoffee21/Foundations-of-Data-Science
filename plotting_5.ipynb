###Pair plot

```
fish = pd.read_csv('/content/drive/My Drive/Data Science/Fish.csv')
fish.head(10)
```

```
f1 = fish[fish['Species'].isin(['Bream','Perch','Pike'])]
```

```
plt.figure(figsize=(11,9))
sns.pairplot(f1)
plt.show()
```

```
sns.pairplot(f1,hue='Species')
```

```
sns.pairplot(f1,hue='Species',palette='rainbow')
```

```
sns.pairplot(f1,hue='Species',palette='rainbow',size=5)
```

```
sns.pairplot(f1,hue='Species',palette='rainbow',size=2,corner=True)
```

```
sns.pairplot(f1,hue='Species', vars=['Length1','Weight','Height','Width'],height=2,aspect=1)
```

```
sns.pairplot(f1,hue='Species', vars=['Length1','Weight','Height','Width'],markers=['o','D','<'],height=2,aspect=1)
```

```
sns.pairplot(f1,hue='Species', x_vars=['Weight','Height','Width'],y_vars=['Length1','Length2','Length3'],markers=['o','D','<'],height=2,aspect=1)
```

```
sns.pairplot(f1,hue='Species', x_vars=['Weight','Height','Width'],y_vars=['Length1','Length2','Length3'],markers=['o','D','<'],kind='reg',height=5, aspect=2)
```

```
sns.pairplot(f1,hue='Species', x_vars=['Weight','Height','Width'],y_vars=['Length1','Length2','Length3'],plot_kws=dict(s=140, linewidth=1,alpha=0.6),height=5, aspect=1)
```

###Heat Maps

```
canada = pd.read_csv('/content/drive/My Drive/Data Science/canada.csv')
canada.head(10)
```

```
canada.drop(columns=['AREA' , 'DEV', 'DevName' , 'REG', 'Type', 'Coverage' , 'AreaName', 'RegName' ], inplace=True)
canada.head()
```

```
canada.rename(columns={'OdName':'Country'} , inplace=True)
canada.set_index(canada.Country,inplace=True)
canada.head()
```

```
canada.index.name=None
canada.head()
```

```
canada = canada.transpose()
canada.head()
```

```
canada1 = canada.loc['2004': , ['Germany' , 'France' , 'Mexico' , 'Bangladesh' ,'Brazil','Israel' , 'Egypt' , 'Colombia' , 'Iraq','Jamaica']]
canada1.head(10)
```

```
plt.figure(figsize=(16,9))
ax=sns.heatmap(canada1)
plt.yticks(rotation=0)
plt.show()
```

```
canada1.isnull().sum()
```

```
canada1
```

```
canada1 = canada1.astype(int)
```

```
canada1.dtypes
```

```
plt.figure(figsize=(16,9))
ax=sns.heatmap(canada1,center=4000)
plt.yticks(rotation=0)
plt.show()
```

```
plt.figure(figsize=(16,9))
ax=sns.heatmap(canada1,vmin=0,vmax=4000)
plt.yticks(rotation=0)
plt.show()
```

```
plt.figure(figsize=(16,9))
ax=sns.heatmap(canada1,vmin=0,vmax=8000,linewidths=0.1)
plt.yticks(rotation=0)
plt.show()
```

```
plt.figure(figsize=(16,9))
ax=sns.heatmap(canada1,vmin=0,vmax=8000,annot=True)
plt.yticks(rotation=0)
plt.show()
```

```
plt.figure(figsize=(16,9))
ax=sns.heatmap(canada1,vmin=0,vmax=8000,annot=True,fmt='d')
plt.yticks(rotation=0)
plt.show()
```

```
plt.figure(figsize=(16,9))
ax=sns.heatmap(canada1,vmin=0,vmax=8000,annot=True,fmt='0.2f')
plt.yticks(rotation=0)
plt.show()
```

```
plt.figure(figsize=(16,9))
ax=sns.heatmap(canada1,vmin=0,vmax=8000,annot=True,fmt='d',annot_kws=dict(size=14,color='#AFFAFF'))
plt.yticks(rotation=0)
plt.show()
```

```
plt.figure(figsize=(16,9))
ax=sns.heatmap(canada1,vmin=0,vmax=8000,annot=True,fmt='d',annot_kws=dict(size=14,color='#AFFAFF'), cbar=False)
plt.yticks(rotation=0)
plt.show()
```

```
plt.figure(figsize=(16,9))
ax=sns.heatmap(canada1,vmin=0,vmax=8000,annot=True,fmt='d',annot_kws=dict(size=14,color='#AFFAFF'), cbar=False,square=True,linewidths=0.2)
plt.yticks(rotation=0)
plt.show()
```
