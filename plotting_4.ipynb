###Dist plot

```
num = np.random.normal(1,10,1000)
sns.distplot(num)
plt.show()
```

```
sns.distplot(num,kde=False)
plt.show()
```

```
sns.distplot(num,kde=False, rug=True)
plt.show()
```

```
sns.distplot(num,kde=True,hist=False,rug=True)
plt.show()
```

```
housing = pd.read_csv('/content/drive/My Drive/Data Science/housing.csv')
housing.head()
```

```
housing.hist(bins=40, figsize=(20,20))
```

```
plt.figure(figsize=(11,7))
sns.distplot(housing['median_income'])
plt.show()
```

```
plt.figure(figsize=(11,7))
sns.distplot(housing['median_income'], bins = 20)
plt.show()
```

```
plt.figure(figsize=(11,7))
sns.distplot(housing['median_income'], bins = 20, hist_kws = dict(edgecolor='#9dab86',linewidth=2), color='g')
plt.show()
```

```
fig, axs = plt.subplots(nrows=2,ncols=2,figsize=(20,20))
sns.distplot(housing['longitude'],color='r',ax=axs[0,0],kde=False,bins=20)
sns.distplot(housing['latitude'],color='g',ax=axs[0,1],kde=False,bins=20)
sns.distplot(housing['housing_median_age'],color='b',ax=axs[1,0],kde=False,bins=20)
sns.distplot(housing['total_rooms'],color='y',ax=axs[1,1],kde=False,bins=20)
```

###Swarm plot

```
ins = pd.read_csv('/content/drive/My Drive/Data Science/insurance.csv')
ins.head()
```

```
plt.figure(figsize=(7,7))
sns.swarmplot(x=ins.smoker,y=ins.charges, hue=ins.sex)
plt.show()
```

```
plt.figure(figsize=(7,7))
sns.swarmplot(y=ins.smoker,x=ins.charges, hue=ins.sex)
plt.show()
```

```
iris=sns.load_dataset("iris")
plt.figure(figsize=(7,7))
sns.swarmplot(x="species",y="petal_width",  data=iris, size=8, linewidth=2,edgecolor="black")
plt.show()
```

###Strip plot

```
unemp = pd.read_excel('/content/drive/My Drive/Data Science/unemployment.xlsx')
unemp.head()
```

```
plt.figure(figsize=(11,8))
sns.stripplot(x=unemp.Unemployed)
plt.show()
```

```
plt.figure(figsize=(11,8))
sns.stripplot(x=unemp.Age,y = unemp.Unemployed, palette="Set1")
plt.show()
```

```
plt.figure(figsize=(11,8))
sns.swarmplot(x=unemp.Age,y = unemp.Unemployed, palette="Set1")
plt.show()
```

```
plt.figure(figsize=(11,8))
sns.scatterplot(x=unemp.Age,y = unemp.Unemployed, palette="Set1")
plt.show()
```

```
plt.figure(figsize=(11,8))
sns.stripplot(x=unemp.Age,y = unemp.Unemployed, palette="Set1", jitter = 0.5)
plt.show()
```
