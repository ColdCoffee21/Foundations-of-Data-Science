###Subplots

```
plt.figure(figsize=(14,6))
x = np.linspace(1,10,100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.subplot(1,2,1)
plt.plot(x,y1)
plt.subplot(1,2,2)
plt.plot(x,y2)
plt.show()
```

```
plt.subplot(2,3,1)
plt.plot(x,y1)
plt.subplot(2,3,4)
plt.plot(x,y2)
plt.show()
```

```
x = np.arange(-50,50)
y1 = np.power(x,2)
y2 = np.power(x,3)
y3 = np.sin(x)
y4 = np.cos(x)
y5 = np.tan(x)
y6 = np.tanh(x)
y7 = np.sinh(x)
y8 = np.cosh(x)
y9 = np.exp(x)

fig1, ax1 = plt.subplots(nrows=3, ncols=3, figsize=(10,10))
ax1[0,0].plot(x,y1,"b-")
ax1[0,0].set_title("Square Function")
ax1[0,0].set_xlabel(r"$X$",fontsize=18)
ax1[0,0].set_ylabel(r"$Y$",fontsize=18)

ax1[0,1].plot(x,y2,"r-")
ax1[0,1].set_title("Cubic Function")
ax1[0,1].set_xlabel(r"$X$",fontsize=18)
ax1[0,1].set_ylabel(r"$Y$",fontsize=18)

ax1[0,2].plot(x,y3,"g-")
ax1[0,2].set_title("Sin Function")
ax1[0,2].set_xlabel(r"$X$",fontsize=18)
ax1[0,2].set_ylabel(r"$Y$",fontsize=18)

ax1[1,0].plot(x,y4,"y-")
ax1[1,0].set_title("Cos Function")
ax1[1,0].set_xlabel(r"$X$",fontsize=18)
ax1[1,0].set_ylabel(r"$Y$",fontsize=18)

ax1[1,1].plot(x,y5,"r-")
ax1[1,1].set_title("Tan Function")
ax1[1,1].set_xlabel(r"$X$",fontsize=18)
ax1[1,1].set_ylabel(r"$Y$",fontsize=18)

ax1[1,2].plot(x,y6,"b-")

ax1[2,0].plot(x,y7,"r-")
ax1[2,1].plot(x,y8,"g-")
ax1[2,2].plot(x,y9,"y-")

```
