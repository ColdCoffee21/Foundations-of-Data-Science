```
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```
columns = ['Age','Workclass','fnlgwt','Education','Education num','Marital Status',
           'Occupation','Relationship','Race','Sex','Capital Gain','Capital Loss',
           'Hours/Week','Native country','Income']
train = pd.read_csv('https://raw.githubusercontent.com/vivek2319/Predicting-US-Census-Income/master/Dataset/adult-training.csv', names=columns)
```

```
train.columns
```

```
train.head()
```

# Q1

```
plt.figure(figsize=(6,4))
sns.scatterplot(train['Capital Loss'],train['Capital Gain'],hue=train['Marital Status'],size=train['Age'])
plt.xlabel("$Capital$ $Loss$",fontsize=12)
plt.ylabel("$Capital$ $Gain$",fontsize=12)
plt.show()
```

Conclusion

Capital loss is more spread in comparison to the captial gain.






# Q2

```
sns.lineplot(x = 'Age',y = 'Hours/Week',data = train,dashes  = False,color = 'green',marker = "<")
sns.despine();
```

Conclusion

Hours/week increases with age till it stabilises at around 40hrs/week after which it reduces.

# Q3

```
sns.boxplot(x=train['Age'], y =train['Marital Status'])
sns.swarmplot(x=train['Age'], y =train['Marital Status'])
plt.show()
```

Conclusion

Displays the ages at which the different maritals status are.

# Q4

```
plt.figure(figsize=(15,7))
sns.violinplot(x=train.Relationship,y=train.Age)
sns.swarmplot(x=train.Relationship,y=train.Age)
```

Conclusion

Displays the spread of each type of relationship status

# Q5

```
plt.figure(figsize=(15,7))
x=sns.boxplot(x='Marital Status',y='Age',hue='Sex',data=train,fliersize=10)
plt.legend(bbox_to_anchor=(1,.5))
x.set_xticklabels(x.get_xticklabels(), rotation=45);
```

Conclusion

Shows comparison between Male and female with respect to age on their various marital statuses.

# Q6

```
plt.figure(figsize=(8,6))
x=sns.violinplot(y='Marital Status',x='Age',split=True,hue='Sex',data=train,cut=4,bw=.05)
plt.title('$Violin$ $Plot$')
plt.legend(bbox_to_anchor=(1.2,1));
plt.show()
```

Conclusion

Spread of various marital statuses with age, with respect to gender.

# Q7

```
fig=sns.barplot(x=train['Income'],y=train['Age'],hue=train['Sex'],dodge=False)
plt.title('Age vs Income',color='white',size=30)
plt.legend(bbox_to_anchor=[1.1, 0.5], loc='center');
plt.show()
```

Conclusion

Shows that males earn more than females in both income ranges, with respect to age.

# Q8

```
g=sns.catplot(x='Age',y='Marital Status',hue='Income',col='Sex',data=train,kind='box')
```

Conclusion

Comparsion of marital status and age, with clear categorical differentiation of gender.

# Q9

```
sns.pointplot(x="Marital Status", y ="Age", hue="Relationship", data=train)
plt.show()
```

Conclusion

Graph of marital status vs age.
