###Pivot, Stack

```
import pandas as pd
```

```
d = pd.DataFrame({
    "Item": ['Item0', 'Item0', 'Item1', 'Item1'],
    "CType":['Gold', 'Bronze', 'Gold', 'Silver'],
    "USD":  ['1$', '2$', '3$', '4$'],
    "EU":  ['1€', '2€', '3€', '4€']
})
```

```
d
```

```
p = d.pivot(index='Item', columns="CType", values = "USD")
p
```

```
print(d[((d.Item=="Item0") & (d.CType=="Gold"))].USD.values)
```

```
print(p[p.index=="Item0"].Gold.values)
```

```
p1 = d.pivot(index='Item', columns="CType")
p1
```

```
d = pd.DataFrame({
    "Item": ['Item0', 'Item0', 'Item0', 'Item1'],
    "CType":['Gold', 'Bronze', 'Gold', 'Silver'],
    "USD":  ['1$', '2$', '3$', '4$'],
    "EU":  ['1€', '2€', '3€', '4€']
})
```

```
d["USD"] = d["USD"].replace('[\$,]', '', regex=True).astype(float)
d["EU"] = d["EU"].replace('[\€,]', '', regex=True).astype(float)
```

###Pivot Table

```
import numpy as np
```

```
p = d.pivot_table(index='Item', columns="CType", values = "USD", aggfunc=np.max)
p
```
